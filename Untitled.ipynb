```python
dataset = pd.concat([dataset1, dataset1], ignore_index=True)
```

```python
del dataset['duration_ms']
del dataset['track_href']
del dataset['uri']
del dataset['playlist_name']
del dataset['analysis_url']
del dataset['track_id']
del dataset['track_album_id']
```

```python
dataset.shape
```

```python
# url = "https://api.deezer.com/track/3135556"
# headers = {"Authorization": f"Bearer {self.token}"}
folder = "previews"

# utw√≥rz folder je≈õli nie istnieje
os.makedirs(folder, exist_ok=True)

for row in dataset.to_dict('records'):
    song_name = row['track_name']
    artist = row['track_artist']
    album = row['track_album_name']
    
    url = f"https://api.deezer.com/search?q=\"{song_name}\" artist:\"{artist}\" album:\"{album}\""
    # url = f"https://api.deezer.com/search?q=DuckTales - From \"DuckTales\" / Extended Version"
    r = requests.get(url)
    data = r.json()
    if data['data']:
        # print(song_name)
        data = data['data'][0]

        # print(data["preview"])

        preview_url = data["preview"]
        r = requests.get(preview_url)
        safe_title = re.sub(r'[\\/*?:"<>|]', "_", f"{song_name} - {artist} - {album}")
        file_path = os.path.join(folder, f"{safe_title}.mp3")
        try:
            with open(file_path, "wb") as f:
                f.write(r.content)
        except:
            print(song_name + ' path_fail')
            with open(os.path.join(folder, "not_finded.txt"), "a", encoding="utf-8") as f:
                f.write(f"{song_name} - {artist} - {album}" + "\n")
    # elif data['data'] == []:
    #     url = f"https://api.deezer.com/search?q=\"{song_name}\""
    #     r = requests.get(url)
    #     data = r.json()
    else:
        print(song_name)
        with open(os.path.join(folder, "not_finded.txt"), "a", encoding="utf-8") as f:
            f.write(f"{song_name} - {artist} - {album}" + "\n")
```
