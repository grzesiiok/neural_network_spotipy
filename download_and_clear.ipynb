```python
import os
import re
import requests

folder = "previews"
os.makedirs(folder, exist_ok=True)

NOT_FOUND_FILE = os.path.join(folder, "not_finded.txt")

def log_not_found(song_name, artist, album):
    print(f"{song_name} - {artist} - {album}<-NOT_FOUND\n")
    with open(NOT_FOUND_FILE, "a", encoding="utf-8") as f:
        f.write(f"{song_name} - {artist} - {album}\n")

def save_preview(song_name, artist, album, preview_url):
    safe_title = re.sub(r'[\\/*?:"<>|]', "_", f"{song_name} - {artist} - {album}")
    file_path = os.path.join(folder, f"{safe_title}.mp3")

    try:
        r = requests.get(preview_url)
        r.raise_for_status()
        with open(file_path, "wb") as f:
            f.write(r.content)
    except Exception:
        print(f"FAILED: {safe_title}")
        log_not_found(song_name, artist, album)

def search_deezer(query):
    r = requests.get(f"https://api.deezer.com/search?q={query}")
    return r.json().get("data", [])

for row in dataset.to_dict("records"):
    song_name = row["track_name"]
    artist = row["track_artist"]
    album = row["track_album_name"]

    # 1Ô∏è‚É£ Pr√≥ba pe≈Çnego dopasowania
    query_full = f"\"{song_name}\" artist:\"{artist}\" album:\"{album}\""
    results = search_deezer(query_full)

    if not results:
        # 2Ô∏è‚É£ Fallback ‚Äì tylko po tytule
        query_title = f"\"{song_name}\""
        results = search_deezer(query_title)

        if not results:
            log_not_found(song_name, artist, album)
            continue

        # pobierz artist/album z API
        api_artist = results[0]["artist"]["name"]
        api_album = results[0]["album"]["title"]
        preview_url = results[0]["preview"]

        save_preview(song_name, api_artist, api_album, preview_url)

    else:
        # idealny match
        preview_url = results[0]["preview"]
        save_preview(song_name, artist, album, preview_url)
```

```python
SPLITER = " - "

with open(Path(os.getcwd()) / 'previews' / 'not_finded.txt', "r", encoding='utf-8') as f:
    lines = f.readlines()
    for line in lines:
        line = line.strip()
        line = re.split(SPLITER, line)
        print(line)
        dataset = dataset.drop(dataset[(dataset['track_name'] == line[0]) & (dataset['track_artist'] == line[1]) & (dataset['track_album_name'] == line[2])].index)
```
